# 음성 파일 -> MFCC 시각화

- sampling rate(sr)
    - 16,000 으로 설정
    - 1초당 몇 번 샘플링(찍어서) 디지털 숫자로 저장할지 정하는 값이다. 
    - 그래서 이거는 1 초를 16,000 조각으로 자라서 각 시점의 진폭을 기록한다.

### 2. 폴더 전체 MFCC 시각화 및 저장

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display

# 입력 폴더(오디오), 출력 폴더(이미지) 설정
input_root = r"D:/project2/vowel_data_noisy/snr_5dB"
output_root = r"D:/project2/snr_mfcc/italian_snr_5db_images"
os.makedirs(output_root, exist_ok=True)

hop_length = 160
n_fft = 400
FIG_SIZE = (15, 10)

for subfolder in os.listdir(input_root):
    subfolder_path = os.path.join(input_root, subfolder)
    if not os.path.isdir(subfolder_path):
        continue
    save_subfolder = os.path.join(output_root, subfolder)
    os.makedirs(save_subfolder, exist_ok=True)
    for fname in os.listdir(subfolder_path):
        if fname.lower().endswith(".wav"):
            file_path = os.path.join(subfolder_path, fname)
            sig, sr = librosa.load(file_path, sr=16000)
            MFCCs = librosa.feature.mfcc(y=sig, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mfcc=13)
            fig = plt.figure(figsize=FIG_SIZE, facecolor='none', edgecolor='none')
            librosa.display.specshow(MFCCs, sr=sr, hop_length=hop_length)
            plt.axis('off')
            save_path = os.path.join(save_subfolder, fname.replace(".wav", ".png"))
            fig.savefig(save_path, bbox_inches='tight', pad_inches=0, transparent=True)
            plt.close(fig)
            print(f"Saved: {save_path}")
```
